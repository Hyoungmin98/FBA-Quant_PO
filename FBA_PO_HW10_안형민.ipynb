# SUMMARY (A STEP BY STEP GUIDE TO THE  BLACK-LITTERMAN MODEL)

※ Black-Litterman Model  

- 직관적이지 않고 고도로 집중된 포트폴리오, 입력 민감도 및 추정 오류 최대화 문제를 극복하는 정교한 포트폴리오 구성 방법 
- 블랙-리터만 모형은 투자자의 믿음을 반영하는 특성으로 인해 개인화된 포트폴리오 구성에 유용하게 사용  
- 투자자들이 직관적이고 다양한 포트폴리오를 초래하는 방식으로 다양한 자산의 성과에 대한 독특한 견해를 시장 평형과 결합할 수 있게 해준다.  
## Section 1  

- 블랙 리터만 모델은 기대수익률 벡터 전체에 오류를 확산시킴으로써 추정 오류 최대화의 문제를 대부분 완화한다.  

- 포트폴리오의 자산 중 하나의 기대 수익률이 약간 증가하는 것이 포트폴리오의 절반 이상의 자산을 포트폴리오에서 제외시킬 수 있음  

expected return → historical return, equal mean return, risk-adjusted equal mean return  
이렇게 대체된 예측값들은 롱&숏 포지션을 가진 포트폴리오에 제약이 없을 때 극단적인 포트폴리오로 이어지게 됨  

※ Equilbrium returns  




max $w'\mu - \frac{\lambda w' \sum w}{2} $  
$w = (\lambda \sum)^{-1}\mu$  

위 공식을 통해 table1의 값들로 세 개의 포트폴리오에 대한 최적의 가중치를 구할 수 있음  


투자자들은 암시적 평형수익과 다른 포트폴리오의 일부 자산의 기대수익률에 대한 구체적인 견해를 가지고 있음.  

※ Example  

View1: International Developed Equity가 절대적인 초과수익률 5.25%를 가질 것이다.(신뢰도: 25%)  
View2: International Bonds가 US Bonds를 25 basis points 능가할 것이다.(신뢰도: 50%)  
View3: US Large Growth와 US Small Growth는 US Large Value와 US Small Value를 2% 능가할 것이다.(신뢰도: 65%)  

- View1은 절대적 관점의 예시이며 View2,3의 경우에는 상대적 관점의 예시  

- 일반적으로 견해가 가중 평균 암시적 평형수익보다 크면 모델은 성과가 뛰어난 자산을 overweighting하는 경향이 있음.  




Q: View Vector  
$\Omega$: diagonal covariance matrix  
(모델의 견해가 서로 독립적이라고 가정하기 때문에 비대각선 원소들은 모두 0)  


- 각 행은 View의 값들을 의미  
- 상대적인 View의 경우 각 행의 값들의 합은 0(실적 우수: +, 실적 저조: -)  

- Satchell and Scowcrofit  
가중치를 동등하게 계산  
→ 비교적 작은 자산들에 큰 변화를 일으키면서 불필요한 오류를 발생시킬 수 있음.  

- Market capitalization weighting scheme  
  


- 투자자가 표현된 견해에 대한 confidence가 작다면 new return vector는 Implied Equilibrium Return Vector에 더 가까워져야 한다.  
- 스칼라는 Implied Equilibrium Return Vector에 주어진 상대적인 가중치에 다소 반비례한다.  

  
- 스칼라 값의 변화는 새로운 결합수익 벡터 E[R]에 영향을 주지 않음(weight/scalar값이 모델에 들어가기 때문)  


※ 블랙리터만 모델의 직관성은 공매도 등 여러 제약이 추가될 때 정확성이 떨어지기에 제약이 있을 경우 MVO에 input에 E[R]을 넣어야함  

- Q행렬의 가중치 초기 설정할 때는 information ratio 가 2.0을 넘지 않도록 하는 것이 좋음. 만약 2 이상일 경우 scalar값만 줄이기

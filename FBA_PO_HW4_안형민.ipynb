chap 7.2

$r_p = w_Dr_D + w_Er_E$  (D: 채권, E: 주식, $w_D = $채권에 할당된 비중, $w_E = $주식에 할당된 비중)  
→ $E(r_p) = w_DE(r_D) + w_EE(r_E)$  
→ $\sigma^2_p = w^2_D\sigma^2_D + w^2_E\sigma^2_E + 2w_Dw_ECov(r_D,r_E)$

$Cov(r_D,r_E) = \rho_{DE}\sigma_D\sigma_E$  
→ $\sigma^2_p = w^2_D\sigma^2_D + w^2_E\sigma^2_E + 2w_Dw_E\rho_{DE}\sigma_D\sigma_E$  
이때 $\rho_{DE} = 1$ 이라면 $\sigma_p = w_D\sigma_D + w_E\sigma_E$  
 $\rho_{DE} = -1$ 이라면 $\sigma_p = \vert{w_D\sigma_D + w_E\sigma_E}\vert $  
그리고 이 표준편차를 0으로 만드는 가중치는 각각 $w_D = \frac{\sigma_E}{\sigma_D + \sigma_E}, w_E = \frac{\sigma_D}{\sigma_D + \sigma_E}$

※ $w_D > 1$ or $w_E < 0$ 일 경우에는 주식을 공매도하고 공매도 수익금을 채권에 투자하는 전략  
   $w_E > 1$ or $w_D < 0$ 일 경우에는 채권을 공매도하고 공매도 수익금을 주식에 투자하는 전략
   
※ minimum-variance protfolio  
$W_{Min}(D) = \frac{\sigma^2_E - Cov(r_D,r_E)}{\sigma^2_D + \sigma^2_E -2Cov(r_D,r_E)}$  
- correlation이 낮을수록 다각화의 잠재적 이익이 더 커진다.

chap 7.4  

- 고객에 따라 최적의 risky portfolio는 배당 수익률 요구사항, 세금 고려사항, 고객 선호사항 등의 제약으로 인해 달라질 수 있다.  
- risky portfolio의 변동성 공식  
$\sigma^2_p = \sum^n_{i=1}\sum^n_{j=1}w_iw_jCov(r_i,r_j)$  
- 가중치가 동등하게 부여된 포트폴리오: $w_i = \frac{1}{n}$  
$\sigma^2_p = \frac{1}{n}\sum^n_{i=1}\frac{1}{n}\sigma^2_i + \sum^n_{j=1, j\neq i}\sum^n_{i=1}\frac{1}{n^2}Cov(r_i,r_j)$

※ average variance & average covariance

$\bar{\sigma^2} = \frac{1}{n}\sum^n_{i=1}\sigma^2_i$  
$\bar{Cov} = \frac{1}{n(n-1)}\sum^n_{j=1,j\neq i}\sum^n_{i=1}Cov(r_i,r_j)$  
→ $\sigma^2_p = \frac{1}{n}\bar{\sigma^2} + \frac{n-1}{n}\bar{Cov}$

- 포트폴리오의 irreducible risk는 구성증권 수익의 공분산에 영향을 받는다.

※ The Markowitz Model

minimize $\frac{1}{2}\sum^n_{i,j=1}w_iw_j\sigma_{ij}$  
s.t. $\sum^n_{i=1}w_i\bar{r_i} = \bar{i}$  
     $\sum^n_{i=1}w_i = 1$
     

※ Solution of the Markowitz Problem

- using Lagrange multipliers

$\sum^n_{j=1}\sigma_{ij}w_j - \lambda\bar{r_i} = \bar{r}$  
$\sum^n_{i=1}w_i\bar{r_i} = \bar{r}$  
$\sum^n_{i=1}w_i = 1$  

※ Nonnegativity Constraints 

- 공매도 허용

minimize $\frac{1}{2}\sum^n_{i,j=1}w_iw_j\sigma_{ij}$  
s.t. $\sum^n_{i=1}w_i\bar{r_i} = \bar{i}$  
     $\sum^n_{i=1}w_i = 1$  
    (+) $w_i \geq 0$

```python
#(a)


import yfinance as yf

# 시가총액 상위 다섯 개 기업의 심볼(symbol) 리스트
top_five_symbols = ["SMCI", "NVDA", "META", "MSFT", "CRWD"]

# 데이터 다운로드
start_date = "2018-01-01"
end_date = "2023-12-31"
top_five_data = yf.download(top_five_symbols, start=start_date, end=end_date)['Adj Close']

# 다운로드한 데이터를 CSV 파일로 저장
top_five_data.to_csv("top_five_returns.csv")
```

```python
#(b)

# 일일 수익률 계산
daily_returns = top_five_data.pct_change()

# 평균 일일 수익률 계산
average_daily_return = daily_returns.mean()

# 연간 수익률 계산 (평균 일일 수익률 * 252)
average_annual_return = average_daily_return * 252

# 공분산 행렬 계산
cov_matrix_daily = daily_returns.cov()

# 연간 공분산 행렬 계산 (공분산 행렬 * 252)
cov_matrix_annual = cov_matrix_daily * 252

# 결과 출력
print("평균 연간 수익률:")
print(average_annual_return)
print("\n연간 공분산 행렬:")
print(cov_matrix_annual)
```

```python
#(c)

import numpy as np
import matplotlib.pyplot as plt

# 주어진 정보
tickers = ["CRWD", "META", "MSFT", "NVDA", "SMCI"]
mean_returns = [0.504396, 0.204932, 0.303639, 0.517529, 0.615352]
cov_matrix_annual = np.array([
    [0.358067, 0.096067, 0.085015, 0.150871, 0.089815],
    [0.096067, 0.182890, 0.078538, 0.119703, 0.075615],
    [0.085015, 0.078538, 0.090745, 0.107548, 0.060338],
    [0.150871, 0.119703, 0.107548, 0.263891, 0.118655],
    [0.089815, 0.075615, 0.060338, 0.118655, 0.362395]
])

# 몬테카를로 시뮬레이션
num_trials = 20000


for i in range(num_trials):
    # 무작위로 포트폴리오 생성
    weights = np.random.rand(len(tickers))
    weights /= np.sum(weights)

    # 포트폴리오의 연간 수익률과 위험 계산
    portfolio_return = np.sum(mean_returns * weights)
    portfolio_risk = np.sqrt(np.dot(weights.T, np.dot(cov_matrix_annual, weights)))

    # 결과 저장
    results[0, i] = portfolio_risk
    results[1, i] = portfolio_return

# 산점도 그리기
plt.figure(figsize=(10, 6))
plt.scatter(results[0, :], results[1, :])
plt.xlabel('Risk (Standard Deviation)')
plt.ylabel('Return (Average Annual)')
plt.title('Monte Carlo Simulation of Risk and Returns')
plt.grid(True)
plt.show()


```

```python
#(d)

# 무위험 이자율
risk_free_rate = 0.02


excess_returns = results[1, :] - risk_free_rate
portfolio_std = results[0, :]

# 샤프 비율 계산
sharpe_ratios = excess_returns / portfolio_std

# 가장 높은 샤프 비율을 가진 포트폴리오 찾기
max_sharpe_index = np.argmax(sharpe_ratios)
max_sharpe_return = results[1, max_sharpe_index]
max_sharpe_risk = results[0, max_sharpe_index]

# CAL 그리기
plt.figure(figsize=(10, 6))
plt.scatter(results[0, :], results[1, :])
plt.scatter(max_sharpe_risk, max_sharpe_return, color='red', marker='o', s=200, label='Maximum Sharpe Ratio')

plt.plot([0, max_sharpe_risk], [risk_free_rate, max_sharpe_return], color='blue', linestyle='-', linewidth=2, label='Capital Allocation Line (CAL)')

plt.xlabel('Risk (Standard Deviation)')
plt.ylabel('Return (Average Annual)')
plt.title('Capital Allocation Line (CAL) with Maximum Sharpe Ratio')
plt.grid(True)
plt.legend()
plt.show()

```

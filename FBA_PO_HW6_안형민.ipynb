※ The Capital Asset Pricing Model

- Assumption  

1. Individual behavior  
a. Investors are rational, mean-variance optimizers.  
b. Their common planning horizon is a single period.  
c. Investors all use identical input lists, an assumption often termed homogeneous expectations. Homogeneous expectations are consistent with the assumption that all relevant information is publicly available.  

2. Market structure  
a. All assets are publicly held and trade on public exchanges.   
b. Investors can borrow or lend at a common risk-free rate, and they can take short positions on traded securities.  
c. No taxes.  
d. No transaction costs.  

- 각 투자자의 optimal risky portfolio에 기반한 CAL과 CML이 동일  
-> 증권 분석을 건너뛰고 단순히 market portfolio를 보유함으로써 효율적인 포트폴리오를 얻어낼 수 있음  
- 이러한 결과를 mutual fund theorem이라고 한다.  

※ The Risk Premium of the Market Portfolio  

$y = \frac{E(r_M)-r_f}{A\sigma^2_M}$  

simplified CAPM economy($\bar{y} = 1)$  
$E(R_M) = \bar{A}\sigma^2_M$  

   
- SML의 기울기: $E(r_M) - r_f$  
- 저평가된 주식은 SML 위에, 고평과된 주식은 SML 밑에 위치  
  



※ zero-beta portfolio  
$E(r_i) - E(r_Z) = \beta_i[E(r_M) - E(r_Z)]$  
- CAPM에서의 SML의 risk-free rate가 $E(r_Z)$로 대체된 형태  
- $E(r_Z)>r_f$, 즉 SML의 기울기가 더 낮아진다.  

※ equilibrium expected return-beta equation  

$E(R_i) = E(R_M)\frac{Cov(R_i,R_M) + \frac{P_H}{P_M}Cov(R_i,R_H)}{\sigma^2_M + \frac{P_H}{P_H}Cov(R_M,R_H)}$  
($P_H =$ Value of aggregate human capital  , $P_M =$ Market value of traded assets  , $R_M =$ Excess rate of return on aggregate human capital)  

※ Merton's ICAPM expected return-beta equation  

- extra-market risk K를 식별하고 K관련 헤지 포트폴리오를 찾을 수 있다고 가정  
$E(R_i) = \beta_{iM}E(R_M) + \sum^K_{k=1}\beta_{ik}E(R_k)$  

※ A Consumption-Based CAPM  

$E(R_i) = \beta_{iC}RP_C$  
(C: consumption-tracking portfolio, $RP_C$: risk premium associated with consumption uncertainty)  

$RP_C = E(R_C) = E(r_C) - r_f$  

- CCAPM의 market portfolio의 베타값은 1보다 훨씬 크다.  
$E(R_M) = \alpha_M + \beta_{MC}E(R_C) + \epsilon_M$  

CCAPM의 장단점  
- 소비 불확실성과 투자 기회의 특성화를 간결하게 통합할 수 있다.  
- 투자 기회의 특성을 특정 매개변수로 나타내므로 투자자들이 이러한 변화를 쉽게 이해하고 분석할 수 있다. 
- 소비 성장 데이터는 상당한 오차로 측정되며 공개 주기도 길다.(최대 월 1회)

※ Factor Models of Security Returns  

single-factor model  
$R_i = E(R_i) + \beta_iF + e_i$  
($\beta_P = \sum w_i\beta_i$, $E(R_P) = \sum w_iE(R_i)$)  
$\sigma^2_P = \beta^2_P\sigma^2_F + \sigma^2(e_P)$  
$\sigma^2(e_P) = \sum w^2_i\sigma^2(e_i)$  

multifactor models  
- GDP성장과 예상하지 못한 이자 변화율 IR 반영  
$R_i = E(R_i) + \beta_{iGDP}GDP + \beta_{iIR}IR + e_i$  

※ Arbitrage Pricing Theory  
- 자본 예산 책정, 증권 평가 또는 투자 성과평가에 활용하는 수익률에 대한 벤치마크 제공  

Assumption  
1. 증권 수익은 factor model로 설명될 수 있다.  
2. 특이한 위험을 분산시키기에 충분한 증권이 존재한다.  
3. 원활하게 작동하는 증권시장은 차익거래 기회를 허용하지 않는다.  

※ Well-Diversified Portfolios  

$R_P = E(R_P) + \beta_PF + e_P$  → $R_P = E(R_P) + \beta_PF$  

$R_P = \alpha_P + \beta_PR_M$

- APT와 CAPM의 접근방식은 다르나 SML 결과는 동일함. 두 모델 모두 회사별 특정 리스크와 시스템적 리스크 사이의 구별을 강조한다.  

※ A Multifactor APT  

two-factor model  
$R_i = E(R_i) + \beta_{i1}F_1 + \beta_{i2}F_2 + e_i$  

competing portfolio  
$E(r_Q) = \beta_{P1}E(r_1) + \beta_{P2}E(r_2) + (1-\beta_{P1}-\beta_{P2})r_f$  


※ The Fama-French Three-Factor Model  

$R_{it} = \alpha_i + \beta_{iM}R_{Mt}+\beta_{iSMB}SMB_t + \beta_{iHML}HML_t + e_{it}$  

SMB = Small Minus Big  
HML = High Minus Low

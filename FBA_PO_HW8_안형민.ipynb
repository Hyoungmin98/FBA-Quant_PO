※ geometric average

※ Internal rate of return

## Adjusting Returns for Risk  

※ comparison universe



- 같은 universe 내에서 펀드매니저가 집중하는 주식 종목이 다를 수 있고, 수익이 일정하지 않고 불규칙적으로 일어난다는 점에서 오해의 소지가 있음  

## Sharpe ratio  

sharpe ratio = $\frac{E(r_p)-r_f}{\sigma_p}$  

- 수용 가능한 포트폴리오가 되기 위해서는 시장 지수보다 더 높은 샤프 비율을 제공해야 함  
- 포트폴리오의 성과 순위를 매길 순 있지만 수치 해석은 어려움  
- overall risky portfolio를 위해 경쟁하는 포트폴리오의 순위를 매길 때 적용

※ $M^2$ Measure  

$M^2_p = r_{p^*} - r_M$  
($P^*$: adjusted porfolio)  

- active portfolio P를 T-bills와 혼합하여 조정된 포트폴리오가 passive marekt index(S&P 500) 와 동일한 변동성을 가지게된다.  
  → 단순히 수익만을 비교하여 성과 측정 가능  
  
  




## The Treynor Ratio



T-line의 기울기: $T_p = \frac{E(r_p)-r_f}{\beta_p}$  

- Sharpe ratio의 total risk 대신 systematic risk 사용  
- overall risky portfolio를 형성하기 위해 혼합되는 포트폴리오의 순위를 매길 때 적용  


## The Information Ratio  

information ratio: $\frac{\alpha_p}{\sigma(e_p)}$  

- 벤치마크 포트폴리오와 혼합될 포트폴리오를 평가할 때 적용  

투자전략에 따라서 성과측정 방법을 다르게 적용시키는 것이 중요  

※ The Role of Alpha in Performance Measures  

Treynor: $T_p - T_M = \frac{\alpha_p}{\beta_p}$  

Sharpe: $S_p - S_M = \frac{\alpha_p}{\sigma_p} -(1-\rho)S_M$  

Information ratio: $\frac{\alpha_p}{\sigma(e_p)}$  

- 알파 값이 양수여야 market index를 능가할 수 있음  


## SCL Regression  

- 펀드매니저의 실력을 사전에 평가하기 위함

$H_0: \alpha = 0$  

standard error: $\hat{\sigma}(\alpha) = \frac{\hat{\sigma}(e)}{\sqrt{N}}$  

t-statistic: $t(\hat{\alpha}) = \frac{\hat{\alpha}}{\hat{\sigma}(\alpha)} = \frac{\hat{\alpha}\sqrt{N}}{\hat{\sigma}(e)}$  

- 펀드매니저의 실력($\alpha$)을 평가하기 위해서는 N months가 필요하다  
→ N값이 너무 크게 나오게 됨으로 인해 성능 평가가 어려움

## MRAR 

Morningstar risk-adjusted return: MRAR($Y$) = $[\frac{1}{T}\sum^T_{t=1}(\frac{1+r_t}{1+r_{ft}})^{-Y}]^{\frac{12}{Y}}-1$  
(t = 1...,T are monthly observations)  

- 투자 수익률을 해당 자산의 투자 위험 수준에 따라 조정  
- 고위험 고수익 자산의 실제 수익률을 조정하여 투자자에게 보다 정확한 정보를 제공

## Market Timing 

- Treynor and Mazuy  
$ r_p - r_f = a + b(r_M - r_f) + c(r_M-r_f)^2 + e_p  $
(a,b,c는 회귀분석에 의해 추정)  

- Henriksson and Merton  
$ r_p - r_f = a + b(r_M - r_f) + c(r_M-r_f)D + e_p  $  
(Dsms $r_M > r_f$ 일 때 1이고 그렇지 않으면 0)  


※ Valuing Market Timing as a Call Option  

MV(Perfect timer per $\$$ of assets) = C = $2N(1/2\sigma_M \sqrt{T}) - 1$  

MV(Imperfect timer) = $(P_1 + P_2 -1) * C = (P_1 + P_2 -1)[2N(1/2\sigma_M \sqrt{T}) - 1]$  
(timing ability: $P_1 + P_2 -1$)  


## Performance Attribution Procedures  

※ attribution method  
- managed portfolio P 와 benchmark portfolio B(bogey)의 수익 차이를 설명  
- bogey는 포트폴리오 매니저가 완전히 수동적인 전략을 따르면 수익을 얻을 수 있도록 설계되었다.  

$r_B = \sum^n_{i=1}w_{Bi}r_{Bi}$  

$r_P = \sum^n_{i=1}w_{Pi}r_{Pi}$   

$r_P - r_B = \sum^n_{i=1}(w_{Pi}r_{Pi} - w_{Bi}r_{Bi})$  

Contribution from asset allocation = $(w_{Pi}-w_{Bi})r_{Bi}$   
Contribution from security selection = $w_{Pi}(r_{Pi}-r_{Bi})$  

Total contribution from asset class i = Contribution from asset allocation + Contribution from security selection 
$= w_{Pi}r_{Pi} - w_{Bi}r_{Bi}$
